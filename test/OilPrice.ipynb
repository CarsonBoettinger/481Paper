```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import openpyxl
```

```python
# Reading the data
route = 'OilPriceData.xlsx'
df = pd.read_excel(route, engine='openpyxl')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming that df is your DataFrame
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 8))

# Plot each oil price
for column in df.columns[1:]:
    plt.plot(df['Date'], df[column], label=column)

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dfNoDate = df.drop(df.columns[0], axis=1)
# Assuming that df is your DataFrame
corr_matrix = dfNoDate.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin =-1, vmax=1)
plt.show()
```
